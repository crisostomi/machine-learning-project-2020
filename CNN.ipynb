```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import cv2
import os
import tensorflow.compat.v1 as tf
import skimage.transform
import sklearn
import random
from keras import optimizers, models
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, ZeroPadding2D, Dense, Flatten, Dropout
from keras.callbacks import EarlyStopping
from keras.regularizers import l2
from keras.preprocessing.image import ImageDataGenerator
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
```

### Paths

```python
two_breeds_path = './dataset/2-breeds/'
five_breeds_path = './dataset/5-breeds/'
ten_breeds_path = './dataset/10-breeds/'
```

### Functions

```python
img_shape = (150,150,3)

def load_pictures(dataset_path):
    X, y = [], [];
    for breed in os.listdir(dataset_path):
        breed_folder = dataset_path+breed
        for img_file in os.listdir(breed_folder):
            img_path = breed_folder + '/'+ img_file
            img = cv2.imread(img_path)
            img_variations = get_augmented_imgs(img)
            norm_images = normalize_imgs(img_variations)
            X += norm_images
            y += [breed for img in norm_images]
    return np.asarray(X),np.asarray(y)

def get_augmented_imgs(img):
    scaled_out = scale_out(img)
    right_rotated = skimage.transform.rotate(img, angle=45, mode='reflect', preserve_range=True)
    left_rotated = skimage.transform.rotate(img, angle=315, mode='reflect', preserve_range=True)
    flipped = np.fliplr(img)
    translated_right = translate(img, 'right', 0.20)
    translated_left = translate(img, 'left', 0.20)
    noised = add_noise(img)
    aug_imgs = [img, scaled_out, right_rotated, left_rotated, flipped, translated_right, translated_left, noised]
    return aug_imgs

def normalize_imgs(imgs):
    norm_imgs = []
    for img in imgs:
        resized_img = cv2.resize(img, (150,150))
        norm_img = np.clip(resized_img/255.0, 0.0, 1.0)
        norm_imgs.append(norm_img)
    return norm_imgs

def scale_out(img):
    height, width, channels = img.shape
    scaled_out = skimage.transform.rescale(img, scale=(2.0,2.0,1.0), mode='constant', preserve_range=True)
    cropped = scaled_out[int(0.125*2*height):int(0.875*2*height), int(0.125*2*width):int(0.875*2*width)]
    return cropped

def translate(img, direction, amount):
    height, width, channels = img.shape
    pad = int(amount*width)
    pad_left, pad_right, pad_top, pad_bottom = 0,0,0,0
    new_width = width + pad
    if direction == 'right':
        pad_right = pad
        output = cv2.copyMakeBorder( img, pad_top, pad_bottom, pad_left, pad_right, cv2.BORDER_REPLICATE)
        img = img[:, pad:new_width]
    elif direction == 'left':
        pad_left = pad
        output = cv2.copyMakeBorder( img, pad_top, pad_bottom, pad_left, pad_right, cv2.BORDER_REPLICATE)
        img = img[0, pad:width]
    return output

def add_noise(img):
    noise = tf.random_normal(shape=tf.shape(img), mean=0.0, stddev=2.0, dtype=tf.float32)
    output = tf.add(img, noise)
    return np.asarray(output)
```

# 1. Feature Engineering

## 1.1. Load data and data augmentation

```python
X_dataset,y_dataset = load_pictures(two_breeds_path)
```

```python
def plot_imgs_from_dataset(n):
    fig, axes = plt.subplots(n//4, 4, figsize=(15,12))
    rand_int = random.randint(0, len(X_dataset))
    for i in range(n):
        img = X_dataset[rand_int+i]
        b,g,r = cv2.split(img)           
        img = cv2.merge([r,g,b])    
        axes[i//4][i%4].set_xticks([])
        axes[i//4][i%4].set_yticks([])
        axes[i//4][i%4].set_xlabel(y_dataset[rand_int+i])
        axes[i//4][i%4].imshow(img)
    plt.show() 
        
plot_imgs_from_dataset(12)
```

### Split dataset in train and test set

```python
X_dataset, y_dataset = shuffle(X_dataset, y_dataset)
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X_dataset, y_dataset, test_size=0.20, random_state=0)
X_train = X_train.reshape((len(X_train), 150, 150, 3))
X_test = X_test.reshape((len(X_test), 150, 150, 3))
```

### One-hot encoding on labels

```python
le = LabelEncoder()

le.fit(y_train)
y_train = le.transform(y_train) 
le.fit(y_test)
y_test = le.transform(y_test)

print("Check if shapes are correct: \n \t X_train shape: {}, y_train shape: {} \n\t X_test shape: {}, y_test shape: {}".format(X_train.shape, y_train.shape, X_test.shape,y_test.shape))

# convert the array of integer labels to onehot encoding
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
numb_labels = y_train.shape[1]
```

# 2. Model

```python
def create_model():
    model = models.Sequential()
    model.add(Conv2D(16, (3,3), activation='relu', input_shape=img_shape))
    model.add(MaxPooling2D((2, 2), padding='same'))
    model.add(Conv2D(16, (3,3), activation='relu'))
    model.add(MaxPooling2D((2, 2), padding='same'))
    model.add(Conv2D(16, (3,3), activation='relu'))
    model.add(MaxPooling2D((2, 2), padding='same'))
    model.add(Conv2D(16, (3,3), activation='relu'))
    model.add(Flatten())
    model.add(Dropout(0.33))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(numb_labels, activation='softmax'))
    return model
```

```python
model = create_model()
model.summary()
```

### Compile and fit model

```python
opt = optimizers.Adam(learning_rate=1e-5)
early_stop = EarlyStopping(monitor='val_accuracy', mode='max', min_delta=0.005, patience=30, verbose=1)
callbacks = [early_stop]

model.compile(optimizer = opt, loss = 'categorical_crossentropy', metrics = ['accuracy'])

epochs = 100
batch_size = 32
history = model.fit(X_train,y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test,y_test), callbacks=callbacks)
```

# 3. Evaluation

### Plot the accuracy of train vs the accuracy of the test

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.savefig('binary-classification-acc.png')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.savefig('binary-classification-loss.png')
plt.show()
```
